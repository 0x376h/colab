```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers.generation import GenerationConfig
import torch
torch.manual_seed(1234)
```

```python
tokenizer = AutoTokenizer.from_pretrained("4bit/Qwen-VL-Chat-Int4", trust_remote_code=True)
```

```python
model = AutoModelForCausalLM.from_pretrained("4bit/Qwen-VL-Chat-Int4", device_map="cuda", trust_remote_code=True).eval()
```

```python
query = tokenizer.from_list_format([
    {'image': 'https://inews.gtimg.com/om_bt/OxgUZMqqRU-x9D5OvGZxGK9UWBhxdfCQV2m3AQpgbxlSwAA/641'}, # Either a local path or an url
    {'text': '请用200字来十分详细描述下这张图片'},
])
response, history = model.chat(tokenizer, query=query, history=None)
print(response)
```
